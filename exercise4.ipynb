```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,6,8])
y=np.array([2,4,6,9])
plt.plot(x,y)
plt.plot(x,y,marker='*',linestyle='dotted',color='r',ms=20,mfc='g')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,4,8,12])
y=np.array([1,3,6,8])
plt.bar(x,y,width=2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,4,8,12])
y=np.array([1,3,6,8])
plt.hist(x,y,width=2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([150,150,150,150])
y=([0,0.6,0,0])
a=["apple","mango","orange","greenaple"]
c=["red","yellow","orange","green"]
plt.pie(x,labels=a,colors=c,explode=y,shadow=True)
plt.show()
```
